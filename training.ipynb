```python
# Instalar pacotes iniciais

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_histogram
from qiskit.primitives import StatevectorSampler
import math
import matplotlib.pyplot as plt
```

```python
# Respostas possíveis da roleta

responses = ["Sim",                     #000
             "Hoje não",                #001
             "Definitivamente",         #010
             "Tentar novamente",        #011
             "Sinais apontam para sim", #100
             "Provavelmente não",       #101
             "Claro!",                  #110
             "Não parece bom"]          #111
```

```python
# Crie um circuito com 3 qubits

circuit = QuantumCircuit(3)

# Crie uma porta Hadamard nos qubits 0, 1, e 2

circuit.h(0)
circuit.h(1)
circuit.h(2)

# Pesar a possibilidade de um resultado usando porta Ry
circuit.ry(2 * math.asin(1 / math.sqrt(8)), 0)
circuit.ry(2 * math.asin(1 / math.sqrt(8)), 1)

# Adicione a medição no seu circuito

circuit.measure_all()

# Visualize seu circuito antes de executar
circuit.draw()
```

```python
shots = 1

# Invocar o Sampler, executar nosso circuito 1024 vezes para provar sobreposição igual e mostrar resultados
sampler = StatevectorSampler()
pub = (circuit)
job = sampler.run([pub], shots=shots)
result = job.result()
counts = result[0].data.meas.get_counts()
counts

aux = 0

if shots > 1:
  # Acessar a primeira chave que contém o resultado
  for key in range(8):
    if list(counts.values())[key-1] > aux:
      aux = list(counts.values())[key-1]
      first_key = list(counts.keys())[key-1]
else:
  first_key = list(counts.keys())[0]

# Transformando em inteiro
# Resultado é dado na base 2 então precisamos comunicar isso para assumir a base 10
Integer_Value = int(first_key, 2)

# Imprimir resposta da roleta
print(responses[Integer_Value])
```
